```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from scipy import stats
```

## Load and plot values on a graph. Plot it as connected dots(line) and scattered dots.

```python
X, y = make_regression(n_samples=100, n_features=1, noise=.1)
plt.scatter(X,y)
```

## Plot x vs y^2 graph

```python
import numpy as np
y = np.asarray(y)
```

```python
y2=y*y
plt.scatter(X,y2)
# plt.plot(X,y2)
```

## Plot best fit line using stats library.

```python
x = np.random.random(100)
y = np.random.random(100)
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
plt.scatter(x,y)
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
```

## Compare the best fit line with another line joining the two extremes, any two points.

```python
plt.scatter(x,y)
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.plot([x[0],x[99]],[y[0],y[99]])
```

## Change value of y and see the effect of outliers on the line.

```python
y[70:95]=y[70:95]*2
y
plt.scatter(x,y)
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.plot([x[92],x[94]],[y[92],y[94]])
```

## Store every 10th value and find best fit curve.

```python
x_ten=x[::10]
y_ten=y[::10]
slope_ten, intercept_ten, r_value_ten, p_value_ten, std_err_ten = stats.linregress(x_ten, y_ten)
plt.scatter(x,y)
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.plot(x, intercept_ten + slope_ten*x, 'g', label='fitted line')

```
